```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits
data_digits = load_digits()
data_digits
```

```python
X= data_digits.data
y= data_digits.target
```

```python
np.unique(y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
LDA = LinearDiscriminantAnalysis()
LDA_transformed = LDA.fit_transform(X,y)
```

```python
LDA.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.scatter(LDA_transformed[:,0], LDA_transformed[:,1], c=y, cmap='rainbow', edgecolors= 'g')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scale = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scale, y, test_size = 0.2)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 9)
X_train_lda = lda.fit_transform(X_train,y_train)
X_test_lda = lda.transform(X_test)
```

```python
lda.explained_variance_ratio_
```

```python
plt.scatter(X_train_lda[:,0], X_train_lda[:,1], c= y_train, cmap='rainbow', edgecolors= 'g')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_train_lda, y_train, test_size=0.2)

from sklearn.linear_model import LogisticRegression 

obj= LogisticRegression()

obj.fit(X_train, y_train)

predicts = obj.predict(X_test)

from sklearn import metrics
print(metrics.accuracy_score(predicts,y_test))
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
X = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) 
X_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit_transform(X_scaled)

print("pca.explained_variance_ratio__", pca.explained_variance_ratio_)

plt.bar(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_)

var_c = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1,len(var_c)+1), var_c)
```

```python
pca = PCA(0.60)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.25)

from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)
prediction = lr.predict(X_test)
```

```python
from sklearn import metrics 
metrics.accuracy_score(y_test, prediction)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, prediction))
```
