```python
lst = ['Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat']
series = pd.Series(lst)
df_series = pd.DataFrame(series)
print(df_series[0].str.upper())
print(df_series[0].str.lower())
print(len(df_series[0]))
```

```python
count= 0
for i in series:
    if series.empty() == True:
        print(i, 0)
    else:
        for j in range(0, len(i)): 
            count= count+1
        print(i, count)
        count = 0
```

```python
lst = ['12', '-$10', '$10,000']
series = pd.Series(lst)
for i in range(0, series.size):
    for j in range(0, len(series[i])):
        if series[i][j] != '$':
            print(series[i][j])
```
