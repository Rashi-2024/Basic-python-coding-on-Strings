```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Rameshwar Agarwal\Desktop\python\assignment 14\bio-degradabale-data.csv', delimiter=';', header=None)
```

```python
df
```

```python
X=df.iloc[:, 0:41]
y=df.iloc[:, 41]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test= train_test_split(X,y,test_size=0.2)

from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()

fit_tree = tree.fit(X_train, y_train)

predict = tree.predict(X_test)

from sklearn import metrics
print(metrics.accuracy_score(predict,y_test))

from sklearn.model_selection import cross_val_score
print(cross_val_score(fit_tree, X, y, cv=5, scoring='accuracy').mean())
```

```python
from sklearn.ensemble import AdaBoostClassifier

abc = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1))

abc.fit(X_train, y_train)

predict = abc.predict(X_test)

print(metrics.accuracy_score(predict, y_test))
```
