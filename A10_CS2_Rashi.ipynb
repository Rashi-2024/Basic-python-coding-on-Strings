```python
dogs= imread(r'C:\Users\Rameshwar Agarwal\Desktop\python\assignment 10\dogs.jpeg')
print(dogs.shape, dogs.ndim)
```

```python
re_dog= dogs.reshape(-1,3)
print(re_dog.ndim, re_dog.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(re_dog)
```

```python
kmeans.cluster_centers_
```

```python
print (kmeans.labels_)
print (len(kmeans.labels_))
```

```python
#print (type(kmeans.labels_))
#unique, counts = np.unique(kmeans.labels_, return_counts=True)
#print(dict(zip(unique, counts)))
```

```python
segmented_image = kmeans.cluster_centers_[kmeans.labels_]
```

```python
segmented_image
```

```python
segmented_image = segmented_image.reshape(dogs.shape)
segmented_image.shape
```
