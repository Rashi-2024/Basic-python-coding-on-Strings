```python
temp= pd.read_csv(r"C:\Users\Rameshwar Agarwal\Desktop\python\assinment 4\CityTemps.csv")
temp
```

```python
Moscow_temp = np.array(temp["Moscow"])
SanF_temp = np.array(temp["San Francisco"])
```

```python
plt.hist(Moscow_temp, bins= range(-10, 25, 5), ec="black", label = "Moscow")
plt.xlabel("Temp")
plt.ylabel("Frequency")


plt.hist(SanF_temp, bins= range(10, 25, 2), ec="red", alpha = 0.7, label="San Francisco")
plt.xlabel("Temp")
plt.ylabel("Frequency")
plt.title("Temp variation over period of 2 years, 2014-2015")
plt.legend(loc = "upper right")
```

```python
X = [1,2,3,4]
Y = [20, 21, 20.5, 20.8]
```

```python
plt.plot(X, Y, marker = "o", mfc = "blue")

plt.axis([0, 5, 20, 21]) 

plt.xlabel("X", fontsize =20)
plt.ylabel("Y")
plt.title("X v/s Y")

error = [0.12, 0.13, 0.2, 0.1]
plt.errorbar(X,Y, yerr = error, fmt=" ")

plt.plot(figsize=(4,5))
plt.rcParams["figure.dpi"] = 100
```
