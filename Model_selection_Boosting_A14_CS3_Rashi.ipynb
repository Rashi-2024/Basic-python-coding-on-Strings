```python
df.rename(columns={'Unnamed: 5':'onpix', 
                   'onpix':'x-bar',
                   'x-bar':'y-bar', 
                   'y-bar':'x2bar', 
                   'x2bar':'y2bar', 
                   'y2bar':'xybar', 
                   'xybar':'x2ybr', 
                   'x2ybr':'xy2br', 
                   'xy2br':'x-ege',
                   'x-ege':'xegvy',
                   'xegvy':'y-ege', 
                   'y-ege':'yegvx',
                   'yegvx':'A'}, inplace=True)
```

```python
df.drop(labels= ['A', 'Unnamed: 18'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
for i in range(1,17):
    X= df.iloc[:, 1:(i+1)]
    y=df.iloc[:, 16]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    abc = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), n_estimators=100)
    abc.fit(X_train, y_train)
    predict = abc.predict(X_test)
    print(i, "-->", metrics.accuracy_score(y_test, predict), "\n")
```

```python
abc1 = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=2), n_estimators=100)
abc1.fit(X_train, y_train)
predict = abc1.predict(X_test)
metrics.accuracy_score(y_test, predict)
```
