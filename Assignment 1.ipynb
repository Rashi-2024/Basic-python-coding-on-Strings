```python
print("enter a number - the program will check whether it is palindrome or not")
num = input()
j = len(num)-1
if (len(num) % 2 == 1):
    for i in range (0, len(num)):  
        if num[i] == num[j]:
            if i == j:
                print ("palindrome number")
                break
            j = j-1
        else:
            print ("not palindrome number")
            break
else:
    for i in range (0, len(num)):  
        if num[i] == num[j]:
            if i == j + 1:
                print ("palindrome number")
                break
            j = j-1
        else:
            print ("not palindrome number")
            break
```
